```python
import numpy as np 
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
import pandas as pd
from IPython.display import Image, display
import os
```

```python
csv_filepath = "/home/x_lab/workspace/roboracer/src/gp-ws/training_results.csv"
data = pd.read_csv(csv_filepath)

lr_instance = 0.01
inducing_instance = 800
independent_instance = True
model_type_instance = 1
epochs_instance = 20
if_norm_instance = True

matching_rows = data[
    (data['lr'] == lr_instance) &
    (data['inducing'] == inducing_instance) &
    (data['independent'] == independent_instance) &
    (data['model_type'] == model_type_instance) &
    (data['epoch'] == epochs_instance) &
    (data['if_norm'] == if_norm_instance)
]

print(matching_rows)

index = -1

eval_path = data[
    (data['lr'] == lr_instance) &
    (data['inducing'] == inducing_instance) &
    (data['independent'] == independent_instance) &
    (data['model_type'] == model_type_instance) &
    (data['epoch'] == epochs_instance) &
    (data['if_norm'] == if_norm_instance)
]['eval_dir'].values[index]

saved_path = data[
    (data['lr'] == lr_instance) &
    (data['inducing'] == inducing_instance) &
    (data['independent'] == independent_instance) &
    (data['model_type'] == model_type_instance) &
    (data['epoch'] == epochs_instance) &
    (data['if_norm'] == if_norm_instance)
]['model_dir'].values[index]

print(f"Evaluation path for the specified parameters: {eval_path}")

# Find the first PNG file in the eval_path directory
png_files = [f for f in os.listdir(eval_path) if f.endswith('.png')]
if png_files:
    display(Image(filename=os.path.join(eval_path, png_files[1])))
else:
    print("No PNG files found in the evaluation path.")

if 'training_loss.png' in png_files:
    display(Image(filename=os.path.join(eval_path, 'training_loss.png')))
else:
    print("training_loss.png not found in the evaluation path.")
```

```python
training_time = matching_rows['training_time_sec'].values[index]
print(f"Training time for the specified parameters: {training_time} seconds")
prediction_time = matching_rows['eval_time_sec'].values[index]
print(f"Prediction time for the specified parameters: {prediction_time} seconds")
```
